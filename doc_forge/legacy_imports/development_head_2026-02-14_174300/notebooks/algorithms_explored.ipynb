### 4.1 INSERT
- **Data Structure**: Singly linked list (or other set representation).

```
INSERT(L, x):
   x.next ‚Üê L.head
   L.head ‚Üê x
```

### 4.2 DELETE
- **Data Structure**: Singly linked list.

```
DELETE(L, x):
   if L.head = x:
      L.head ‚Üê x.next
   else:
      find node y such that y.next = x
      y.next = x.next
```
(Worst-case \( \Theta(n) \) because we must find predecessor.)

### 4.3 UNION
- **Data Structure**: Linked lists representing sets.

```
UNION(L1, L2):
   // If disjoint sets, we can combine in O(1) by linking heads/tails:
   if L1.tail not connected to L2:
      L1.tail.next = L2.head
   return L1
```

## 7. **Maximum Flow Algorithms**

### 7.1 Ford-Fulkerson Method
- **Data Structure**: Flow network \( G = (V, E) \) with capacities \( c(u,v) \).

```
FORD-FULKERSON(G, s, t):
   for each edge (u,v):
      f(u,v) ‚Üê 0
   while there exists an augmenting path p from s to t in residual network G_f:
      cf ‚Üê min{ c_f(u,v) : (u,v) in p }
      for each (u,v) in p:
         f(u,v) ‚Üê f(u,v) + cf
         f(v,u) ‚Üê f(v,u) - cf
   return f
```

### 7.2 MAX-FLOW-BY-SCALING

- Find augmenting paths of capacity at least \( k \).  
- Halve \( k \) repeatedly until \( k = 1 \).

## 9. **Dynamic Programming Algorithms**

### 9.1 MATRIX-CHAIN-ORDER
- **Data Structure**: Dimensions array \( p \) of length \( n+1 \); DP tables \( m \) and \( s \).

```
MATRIX-CHAIN-ORDER(p):
   n ‚Üê length(p)-1
   for i ‚Üê 1 to n:
       m[i,i] = 0
   for l ‚Üê 2 to n: // length of chain
       for i ‚Üê 1 to n-l+1:
           j ‚Üê i + l - 1
           m[i,j] ‚Üê ‚àû
           for k ‚Üê i to j-1:
               q = m[i,k] + m[k+1,j] + p[i-1]*p[k]*p[j]
               if q < m[i,j]:
                  m[i,j] = q
                  s[i,j] = k
   return m, s
```

## 15. **LUP Decomposition**

### 15.1 LUP-DECOMPOSITION
- **Data Structure**: Matrix \( A \), permutation array \( \pi \).

```
LUP-DECOMPOSITION(A):
   n = rows(A)
   for i = 1 to n:
      œÄ[i] = i
   for k = 1 to n:
      p = 0
      for i = k to n:
         if |A[i,k]| > p:
            p = |A[i,k]|
            k' = i
      if p = 0:
         error("singular matrix")
      swap œÄ[k], œÄ[k']
      swap A[k,*], A[k',*]
      for i = k+1 to n:
         A[i,k] = A[i,k] / A[k,k]
         for j = k+1 to n:
            A[i,j] = A[i,j] - A[i,k]*A[k,j]
```

### 15.2 LUP-SOLVE
- **Data Structure**: Decomposed matrix \( A \) (with L and U in place), permutation array \( \pi \).

```
LUP-SOLVE(A, œÄ, b):
   n = length(b)
   // Forward substitution
   for i = 1 to n:
      x[i] = b[œÄ[i]]
      for j = 1 to i-1:
         x[i] = x[i] - A[i,j]*x[j]
   // Back substitution
   for i = n downto 1:
      for j = i+1 to n:
         x[i] = x[i] - A[i,j]*x[j]
      x[i] = x[i]/A[i,i]
   return x
```
